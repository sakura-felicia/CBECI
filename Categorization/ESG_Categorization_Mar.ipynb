```
### import packages
import pandas as pd
import numpy as np
### show all columns
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```
### Import data
df_Mar= pd.read_csv("https://raw.githubusercontent.com/sunshineluyao/ESG_with_benchmark/main/rawdata/SP1500_Mar.csv")
df_Mar.head()
```

```
df_Mar=df_Mar.dropna()
```

```
df_Mar=df_Mar.set_index('Symbol',drop=True)
df_Mar.head()
```

```
df_Mar.index.name=None
df_Mar.head()
```

```
### formating data so that it can be calculated
df_Mar[['Size','Common Shares Outstdg','MktVal Co']]=df_Mar[['Size','Common Shares Outstdg','MktVal Co']].apply(lambda x: x.str.replace(',', ''), axis=1)
df_Mar.head()
```

```
df_Mar[['Size', 'Book Value Per Shr Tangble','Common Shares Outstdg','MktVal Co','Momentum']] = df_Mar[['Size', 'Book Value Per Shr Tangble','Common Shares Outstdg','MktVal Co','Momentum']].astype('float')
df_Mar.head()
```

```
### Generate the new variable:BM_Ratio
df_Mar['BM_Ratio'] = df_Mar['Book Value Per Shr Tangble']*df_Mar['Common Shares Outstdg']/df_Mar['MktVal Co']
df_Mar.head()
```

```
### Generate the categories by quintile cut and ESG rating
df_Mar['Size_Index']=pd.qcut(df_Mar['Size'],5,labels=np.arange(1,6,1))
df_Mar['Momentum_Index']=pd.qcut(df_Mar['Momentum'],5,labels=np.arange(1,6,1))
df_Mar['BM_Ratio_Index']=pd.qcut(df_Mar['BM_Ratio'],5,labels=np.arange(1,6,1))
df_Mar['ESG_Index']=np.where(df_Mar['MSCI ESG Rtg Overall Company Rating'].isin(['A','AA','AAA']), 'High','Low')
df_Mar.head()
```

```
df_Mar_Overview=df_Mar[['Size_Index','Momentum_Index','BM_Ratio_Index','ESG_Index','Monthly Return']]
```

```
df_Mar_Overview_Benchmark = df_Mar_Overview.groupby(by=['Size_Index','Momentum_Index','BM_Ratio_Index',],group_keys=True,sort=True,as_index=False).mean('Monthly Return')
```

```
df_Mar_Overview_Benchmark.head()
```

```
df_Mar_Overview_ESG_High = df_Mar_Overview[df_Mar_Overview['ESG_Index']=='High']
df_Mar_Overview_ESG_High.head()
```

```
df_Mar_Overveiw_ESG_High_Mean = df_Mar_Overview_ESG_High.groupby(by=['Size_Index','Momentum_Index','BM_Ratio_Index',],group_keys=True,sort=True,as_index=False).mean('Monthly Return')
```

```
df_Mar_Overveiw_ESG_High_Mean.head()
```

```
df_Mar_Overview_Benchmark['ESG_High_Return']= df_Mar_Overveiw_ESG_High_Mean['Monthly Return']
df_Mar_Overview_Benchmark.head()
```

```
df_Mar_Overview_Benchmark['Abnormal Return']=df_Mar_Overview_Benchmark['ESG_High_Return']-df_Mar_Overview_Benchmark['Monthly Return']
df_Mar_Overview_Benchmark.head()
```

```
df_Mar_Overview_Benchmark['Month']='Mar'
df_Mar_Overview_Benchmark.head()
```
